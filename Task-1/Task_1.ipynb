### Analyse the data properly and train the most suitable machine learning to analyse if the person is diabetic.

```python
from sklearn.model_selection import train_test_split

# defining dependent and independent variable
x = df.drop("Diabetes_012", axis=1)
y = df["Diabetes_012"]
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.70,random_state=12)
```

```python
#Many models can be used for binary classification of this imbalanced and large dataset
#Using Logistic Regression
import warnings #avoid warning flash
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```

### Test the trained model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV
test1=accuracy_score(y_pred, y_test)
test2=precision_score(y_pred, y_test)
test3=recall_score(y_pred, y_test)
test4=f1_score(y_pred, y_test)
print("Accuracy Score:", test1)
print("Precision Score:", test2)
print("Recall Score:", test3)
print("f1 Score:", test4)
```

### Evaluate the model and tune the model to maximize the accuracy

```python
# Define a python function to compute training & testing set metrics
# The weighted average is taken in order to account for class imbalance
def measure_error(y_true, y_pred, label):
    return pd.Series({'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred, average='weighted'),
        'Recall': recall_score(y_true, y_pred, average='weighted'),
        'f1': f1_score(y_true, y_pred, average='weighted')}, name=label)
# Function to evaluate the model - fits the model, then computes training vs testing set metrics
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Computes the metrics for the training & testing set, and presents them in a dataframe
    train_test_error = pd.concat([measure_error(y_train, y_train_pred, 'Train'),
                                    measure_error(y_test, y_test_pred, 'Test')],
                                    axis=1)
    print(train_test_error, "\n")
# Evaluate the model
evaluate_model(lr, x_train, y_train, x_test, y_test)
```
