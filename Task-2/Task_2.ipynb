### Analyse the data and train a model that divides this data into multiple groups.

```python
#Scaling the data using minmax method: values will be in range 1-10
df=((df - df.min())/(df.max() - df.min())) * 9 +1
df.head()
#repeat steps 3 and 4 until centroids stop scaling
```

```python
# Initialize random centroids
def random_centroids(df, k):
    centroids=[]
    for i in range(k): #k=no. of centroids
        centroid= df.apply(lambda x: float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1)
c=random_centroids(df, 3)
c
```

```python
#label each data point
def get_labels(df, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((df - x) ** 2).sum(axis=1))) #distance formula
    return distances.idxmin(axis=1)
```

```python
#update centroids: by geometric mean
def new_centroids(df, labels, k):
    return df.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

```python
#visualizing 
from sklearn.decomposition import PCA
from IPython.display import clear_output
def plot_clusters(df, labels, centroids, iteration):
    pca = PCA(n_components=2)
    df_2d = pca.fit_transform(df)
    centroids_2d = pca.transform(centroids.T)
    clear_output(wait=True)
    plt.title(f"Iteration {iteration}")
    plt.scatter(x=df_2d[:, 0], y=df_2d[:, 1], c=labels)
    plt.scatter(x=centroids_2d[:, 0], y=centroids_2d[:, 1])
    plt.show()
```

```python
#sample of k-means clustering trained model  with 3 clusters
max_iterations = 100.0
k = 3 #random value for clusters
centroids = random_centroids(df, k)
old_centroids = pd.DataFrame()
iteration = 1
while iteration < max_iterations and not centroids.equals(old_centroids):
    old_centroids = centroids
    labels = get_labels(df, centroids)
    centroids = new_centroids(df, labels, k)
    plot_clusters(df, labels, centroids, iteration)
    iteration = iteration + 1
```

### Determine what would be the best number of groups we should divide the data.

```python
#using elbow method to determine the optimal value of clusters: from the graph it is clear that the optimal value of clusters is 3
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
wcss=[]
X=df.iloc[:,:].values
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init="k-means++", n_init=10, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.xlabel("Numbers of Clusters")
plt.ylabel("WCSS")
plt.show()
```

### Take two columns with the highest correlation and plot a graph between those columns to display the group formation.

```python
# Find the correlation matrix
correlation_matrix = df.corr()
#visualizing correlation_matrix
sns.heatmap(correlation_matrix)

# Find the pair of columns with the highest correlation also by heatmap it can be seen "Total_Phenols" and "Flavanoids" has the highest correlation
max_corr = correlation_matrix.abs().unstack().sort_values(ascending=False)
max_corr = max_corr[max_corr < 1]  # Exclude self-correlation
column1, column2 = max_corr.index[0]

# Plot the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df[column1], y=df[column2])
plt.title(f'Scatter Plot between {column1} and {column2}')
plt.xlabel(column1)
plt.ylabel(column2)
plt.show()

```
